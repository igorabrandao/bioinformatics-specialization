THE CODE LOGIC BEGINS HERE!

```
# ======================================
# precompute_max_prefix_suffix function
# ======================================

# Returns an array such that:
# - len(result) == len(needle) + 1
# - result[k] is the length of the longest prefix of needle[0:k], not counting
#   needle[0:k] itself, that's also a suffix of needle[0:k]
def precompute_max_prefix_suffix(needle):
  result = [0] * (len(needle) + 1)
  # We start from 2 because we always have result[0] == result[1] == 0
  for k in range(2, len(needle) + 1):
    l_minus_one = result[k-1]
    while l_minus_one > 0 and needle[l_minus_one] != needle[k-1]:
      # Repeated "calls to max_prefix_suffix" become repeated array lookups
      l_minus_one = result[l_minus_one]
    if needle[l_minus_one] == needle[k-1]:
      result[k] = l_minus_one + 1
  return result

# ======================================
# match_len function
# ======================================

# Given that needle[0:overlap_region] == haystack[i:i+overlap_region], returns
# the largest k such that needle[0:k] == haystack[i:i+k]
def match_len(needle, haystack, i, overlap_region):
  for j in range(overlap_region, len(needle)):
    if needle[j] != haystack[i+j]:
      return j
  return len(needle)

# ======================================
# PatternMamount_to_advanceatching function
# ======================================
def amount_to_advance(max_prefix_suffix, k):
  if k == 0:
    return 1
  return k - max_prefix_suffix[k]
```

```
# ======================================
# PatternMatching function
# ======================================
def PatternMatching(needle, haystack):
  result = []
  max_prefix_suffix = precompute_max_prefix_suffix(needle)
  i = 0
  overlap_region = 0
  while i <= len(haystack) - len(needle):
    k = match_len(needle, haystack, i, overlap_region)
    if k == len(needle):
      result.append(i)
    overlap_region = max_prefix_suffix[k]
    i += amount_to_advance(max_prefix_suffix, k)
  return result
```
