```
# ======================================
# PatternCount function
# ======================================
def PatternCount(input_, pattern_):
  occurrences = 0
  pattern_size = len(pattern_)

  for idx, chr in enumerate(input_):
    if input_[idx:(idx+pattern_size)] == pattern_:
      occurrences+=1
  return occurrences
```

```
# ======================================
# FrequentWords function
# complexity is O((|Text|^2) * k)
# ======================================
def FrequentWords(input_, patternSize_):
  # Initialize the empty most frequency patterns list
  frequent_patterns = []
  patterns_count = []
  input_size = len(input_)

  # Limit is the |Text| − k
  limit = input_size - patternSize_

  # Loop 1: Account the patterns occurences
  for idx in range(limit):
    # Get the current k-mer
    current_pattern = input_[idx:(idx+patternSize_)]

    # Count the current pattern occurrences and store it
    patterns_count.append(PatternCount(input_, current_pattern))

  # Get the max occurrence
  max_patterns_count = max(patterns_count)

  # Loop 2: Find the most frequent pattern
  for idx in range(limit):
    # Check if the current pattern is the most frequent
    if patterns_count[idx] == max_patterns_count:
      # If so add it to the most frequent pattern list
      frequent_patterns.append(input_[idx:(idx+patternSize_)])

  # Remove duplicates from the most frequent pattern list
  frequent_patterns = list(dict.fromkeys(frequent_patterns))

  # Return the most frequent pattern list
  return frequent_patterns
```
