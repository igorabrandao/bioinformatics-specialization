```
# ======================================
# Skew function
# ======================================
def Skew(input_):
  # Define the skew list with 0 value
  skewList = [0]

  # Define the equivalence list of values
  skewValues = {'A': 0, 'T': 0, 'C': -1, 'G': 1}

  # Iterate over the genome and computate each skew
  for idx, chr in enumerate(input_):
    # Calculate the current skew and add it to the skew list
    skewList.append(skewList[idx] + skewValues[chr])

  # Return the skew list
  return skewList
```

```
# ======================================
# Test section
# ======================================
# Test 1
output = Skew("CATGGGCATCGGCCATACGCC")
print(*output, sep=' ')

print("\n")

# Test 2
output2 = Skew("GAGCCACCGCGATA")
print(*output2, sep=' ')
```
