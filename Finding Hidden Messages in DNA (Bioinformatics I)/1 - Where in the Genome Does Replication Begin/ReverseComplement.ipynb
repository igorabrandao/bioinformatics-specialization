THE CODE LOGIC BEGINS HERE!

```
# ======================================
# GetComplement function
# ======================================
def GetComplement(input_):
  input_ = input_.upper()
  result = ""

  # Iterate over the input string
  for idx, chr in enumerate(input_):
    # Check the lexicographic order
    switcher = {
      "A": "T",
      "C": "G",
      "G": "C",
      "T": "A"
    }
    current_result = switcher.get(input_[idx], "Invalid base")

    # Verify if the current base is valid!
    if (current_result == "Invalid base"):
      return -1
    else:
      result += current_result

  # Return the final result
  return result

# ======================================
# ReverseComplement function
# ======================================
def ReverseComplement(input_):
  # Get the complement nucleotides
  complement = GetComplement(input_)

  # Return the complement reverse
  return complement[::-1]
```
