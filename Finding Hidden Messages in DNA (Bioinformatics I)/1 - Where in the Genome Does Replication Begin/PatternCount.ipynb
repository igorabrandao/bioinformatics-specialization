THE CODE LOGIC BEGINS HERE!

```
# Define the PatternCount function
def PatternCount(input_, pattern_):
  occurrences = 0
  pattern_size = len(pattern_)

  for idx, chr in enumerate(input_):
    if input_[idx:(idx+pattern_size)] == pattern_:
      occurrences+=1
  return occurrences

# Test the PatternCount function
output = PatternCount(input, "TTGCGTCTT")

# Print the output (minus the input in the second line of file)
print(output - 1)
```
